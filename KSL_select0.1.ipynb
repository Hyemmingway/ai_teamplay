```python
data = []
for i in Class_name:
    path = os.path.join(directory, i)
    file_list = os.listdir(path)
    
    for j in file_list:
        name = j
        label1 = i
        label2 = i[3:5]
    
        eda = [name, label1, label2]
        data.append(eda)
```
